```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

%matplotlib inline
```

## Load Data

```python
df = pd.read_csv("./analysis/Army_FY_2020.csv")
```

## Member vs. Witness Speaking Time by Topic

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=df, x="topic_id", y="minutes", hue="role")
```

## Speaking Time Density by Role

```python
sns.displot(data=df, x="minutes", hue="role", kind="kde")
```

## Member vs. Witness Polarity by Topic     

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=df, x="topic_id", y="polarity", hue="role")
```

## Mean Speaking Time by District

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=df, x="district", y="minutes")
```

```python
plt.figure(figsize=(16, 8))
ax = sns.violinplot(data=df, x="district", y="minutes", inner=None, color="1")
ax = sns.stripplot(data=df, x="district", y="minutes")
```

## Mean Polarity by District

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=df, x="district", y="polarity")
```

```python
plt.figure(figsize=(16, 8))
ax = sns.violinplot(data=df, x="district", y="polarity", inner=None, color="1")
ax = sns.stripplot(data=df, x="district", y="polarity")
```

## Speaking Time vs. Polarity by Role

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(data=df, x="minutes", y="polarity", hue="role")
```
